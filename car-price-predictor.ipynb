## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

xlabel = x.copy()
encoders = {}   # store encoders for each column

for col in ['model', 'transmission', 'fuelType']:
    le = LabelEncoder()
    xlabel[col] = le.fit_transform(x[col])  # transform categorical -> numeric
    encoders[col] = le                      # save the encoder
```

## Model training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_one_encoded, y, test_size=0.33, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```
